# 推文数量随时间的变化分析

## 项目概述

本分析旨在探究推文数量随时间的变化情况。通过对推文数据进行时间序列分析，我们可以了解特定时间段内社交媒体的活跃程度。

## 数据处理

首先，我将原始数据中的 `Datetime` 字段转换为 datetime 格式，并将其设置为数据框架的索引。这一步骤对于后续的时间序列分析至关重要。

## 数据分组与计数

接下来，我按小时对数据进行了分组，并计算了每个小时的推文数量。这帮助我们揭示了推文活动随时间的变化模式。

## 可视化展示

使用 seaborn 库绘制了时间序列图，我能够清晰地展现推文数量在不同时间点的变化。图表的大小设定为 15x6 英寸，以便于观察更多细节。

## 结果解读

从图表中，我们可以观察到推文数量在某些小时内出现显著的峰值，这可能表明了社交媒体使用高峰期或特定事件发生的时段。通过进一步分析这些峰值，我们能够对用户行为和事件反应有更深入的理解。


```python
import pandas as pd

# Load the CSV file to examine its contents
file_path = './data/chatgpt1.csv'
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure and types of data it contains
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Converting 'Datetime' to datetime format
data['Datetime'] = pd.to_datetime(data['Datetime'])

# Grouping by date or hour to count the number of tweets
tweet_count_by_time = data.set_index('Datetime').resample('H').size()

# Plotting
plt.figure(figsize=(15, 6))
sns.lineplot(data=tweet_count_by_time)
plt.title('Number of Tweets Over Time')
plt.xlabel('Datetime')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

# 带有"其他"类别的推文语言分布

这张甜甜圈图展示了不同语言推文的分布情况。每个色块代表一个语言类别及其在总推文中的百分比。

- **英语 (en)**: 79.3%
- **日语 (ja)**: 12.5%
- **西班牙语 (es)**: 8.2%
- **其他**: 剩余百分比

主要的推文语言为英语，占据了图表的大部分。日语和西班牙语的推文数量较少，而其他语言的分布没有详细指出。


```python
# To address the issue of the cluttered labels, we will group the languages with small percentages into an 'Other' category
# and only label the more significant segments.

# First, we need to determine which languages to group into 'Other'
other_threshold = 0.05  # Threshold as a percentage (5%)
# Calculate the percentage for each language
tweet_percentages = tweet_count_by_language / tweet_count_by_language.sum()
# Find languages with a percentage lower than the threshold
other_languages = tweet_percentages[tweet_percentages < other_threshold]

# Sum the tweet counts for the 'Other' category
other_count = other_languages.sum()
# Create a new tweet count series without the 'Other' languages
main_languages = tweet_count_by_language.drop(other_languages.index)
# Append the 'Other' category to the main languages
main_languages['Other'] = other_count

# Plotting the pie chart with 'Other' category
plt.figure(figsize=(10, 10))
plt.pie(main_languages, labels=main_languages.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)

# Draw a circle at the center to turn the pie into a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Add title
plt.title('Distribution of Tweets by Language with "Other" Category', pad=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.

# Save the plot as a file
output_file_donut_chart_grouped = './data/donut_chart_with_other.png'
plt.savefig(output_file_donut_chart_grouped)

# Show the plot
plt.show()

# Provide the path to the saved file
output_file_donut_chart_grouped
```

```python
# To understand why there's only one bar in the histograms, we need to inspect the data.
# Let's load the data and look at the statistics for the columns in question to understand their distribution.

# Load the data from the CSV file
data_path = './data/chatgpt1.csv'
data = pd.read_csv(data_path)

# Inspect the statistical summary of the LikeCount, RetweetCount, ReplyCount, and QuoteCount columns
like_count_desc = data['LikeCount'].describe()
retweet_count_desc = data['RetweetCount'].describe()
reply_count_desc = data['ReplyCount'].describe()
quote_count_desc = data['QuoteCount'].describe()

(like_count_desc, retweet_count_desc, reply_count_desc, quote_count_desc)

import numpy as np

# Applying log transformation and adding 1 to avoid log(0)
# We will use np.log1p which is equivalent to np.log(1 + x)
data['LogLikeCount'] = np.log1p(data['LikeCount'])
data['LogRetweetCount'] = np.log1p(data['RetweetCount'])
data['LogReplyCount'] = np.log1p(data['ReplyCount'])
data['LogQuoteCount'] = np.log1p(data['QuoteCount'])

# Plotting histograms for the log-transformed counts
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# LogLikeCount
sns.histplot(data[data['LogLikeCount'] > 0]['LogLikeCount'], bins=30, ax=axes[0, 0], kde=False, color='blue')
axes[0, 0].set_title('Distribution of Log Like Counts')
axes[0, 0].set_xlabel('Log Like Count')
axes[0, 0].set_ylabel('Frequency')

# LogRetweetCount
sns.histplot(data[data['LogRetweetCount'] > 0]['LogRetweetCount'], bins=30, ax=axes[0, 1], kde=False, color='green')
axes[0, 1].set_title('Distribution of Log Retweet Counts')
axes[0, 1].set_xlabel('Log Retweet Count')
axes[0, 1].set_ylabel('Frequency')

# LogReplyCount
sns.histplot(data[data['LogReplyCount'] > 0]['LogReplyCount'], bins=30, ax=axes[1, 0], kde=False, color='red')
axes[1, 0].set_title('Distribution of Log Reply Counts')
axes[1, 0].set_xlabel('Log Reply Count')
axes[1, 0].set_ylabel('Frequency')

# LogQuoteCount
sns.histplot(data[data['LogQuoteCount'] > 0]['LogQuoteCount'], bins=30, ax=axes[1, 1], kde=False, color='purple')
axes[1, 1].set_title('Distribution of Log Quote Counts')
axes[1, 1].set_xlabel('Log Quote Count')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# 推文来源设备的分布

这个饼图展示了不同来源设备发送推文的分布比例。

- **Twitter Web App**: 35.6%
- **Twitter for iPhone**: 24.6%
- **Twitter for Android**: 17.9%
- **IFTTT**: 2.8%
- **其他**: 19.1%

每个部分的大小代表了该设备发送推文的比例，颜色不同以区分各个设备。最大的部分是蓝色，代表Twitter网页端应用；其次是橙色和绿色，分别代表iPhone和Android端的Twitter应用；红色部分较小，代表IFTTT；紫色部分代表所有其他设备的总和。


```python
import re
data['Source_Cleaned'] = data['Source'].apply(lambda x: re.findall('>(.*?)<', x)[0] if x else None)

# Counting the number of tweets per device
tweet_count_by_device = data['Source_Cleaned'].value_counts()

# To handle the label crowding, we can group small categories into 'Other'
# Let's keep only the top 4 categories and group the rest
top_categories = tweet_count_by_device.nlargest(4)
others = tweet_count_by_device[~tweet_count_by_device.index.isin(top_categories.index)]
top_categories['Other'] = others.sum()

# Plotting
plt.figure(figsize=(10, 10))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Tweets by Source Device')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Show the plot
plt.show()
```

# 话题标签使用频率

这个水平柱状图展示了数据集中最常见的10个话题标签及其使用频率。

- Y轴列出了话题标签。
- X轴显示了每个标签的使用频率。
- 柱状图的颜色从顶部的深紫色渐变到底部的浅橙色，可能代表了使用频率的不同级别。

```python
from collections import Counter
import ast

# Since the hashtags are in string representation of list, we need to convert them back to list
# Then we flatten the list of lists into a single list containing all hashtags
all_hashtags = [hashtag for sublist in data['hashtag'].dropna().apply(ast.literal_eval) for hashtag in sublist]

# Counting the frequency of each hashtag
hashtag_counts = Counter(all_hashtags)

# Selecting the top 10 hashtags for visualization
top_hashtags = dict(hashtag_counts.most_common(10))

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=list(top_hashtags.values()), y=list(top_hashtags.keys()), palette="rocket")
plt.title('Top 10 Most Common Hashtags')
plt.xlabel('Frequency')
plt.ylabel('Hashtag')
plt.tight_layout()

# Show the plot
plt.show()
```

# 十大最热门推文

这个条形图展示了十条最热门推文及其受欢迎程度。

- Y轴列出了截短后的推文内容，每个推文显示为最多50个字符，如果超过则以省略号结束。
- X轴表示每条推文的受欢迎得分，得分越高，说明推文越受欢迎。
- 不同的颜色代表不同的推文，颜色从上到下由深变浅。
- 图表的标题为“十大最热门推文”。
- 横轴标签为“受欢迎得分”，纵轴标签为“推文”。
- 图表通过`plt.tight_layout()`确保布局紧凑，使所有元素都适当地展示在画布上。


```python
# Creating a popularity score for each tweet by summing up the likes, retweets, replies, and quotes
data['PopularityScore'] = data['LikeCount'] + data['RetweetCount'] + data['ReplyCount'] + data['QuoteCount']

# Sorting the data by the PopularityScore in descending order
popular_tweets = data.sort_values(by='PopularityScore', ascending=False).head(10)

# Selecting relevant columns to display
popular_tweets_display = popular_tweets[['Text', 'Username', 'LikeCount', 'RetweetCount', 'ReplyCount', 'QuoteCount', 'PopularityScore']]

popular_tweets_display.reset_index(drop=True)
```

```python
import textwrap
import unicodedata
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to remove or replace non-ASCII characters
def clean_text(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')

# Clean the tweet text
popular_tweets_display['Text'] = popular_tweets_display['Text'].apply(clean_text)

# Then you can wrap the text as before
popular_tweets_display['Text'] = popular_tweets_display['Text'].str.wrap(50)
popular_tweets_display['Text'] = popular_tweets_display['Text'].apply(lambda x: '\n'.join(textwrap.wrap(x, width=50)))


# Plotting the popularity of the top 10 tweets
plt.figure(figsize=(12, 12))  # 将高度增加以便于更多的空间显示标签


# Using a bar plot to display the popularity score
sns.barplot(x='PopularityScore', y='Text', data=popular_tweets_display, palette="mako")

popular_tweets_display['Text'] = popular_tweets_display['Text'].str.wrap(50)  # Wraps text after 50 characters
popular_tweets_display['Text'] = popular_tweets_display['Text'].apply(lambda x: '\n'.join(textwrap.wrap(x, width=50)))


plt.title('Top 10 Most Popular Tweets')
plt.xlabel('Popularity Score')
plt.ylabel('Tweet')
plt.tight_layout()

# Show the plot
plt.show()
```

# 用户账户类型分析

这个条形图展示了基于用户名关键词分类的用户账户类型及其对应的推文数量。

- X轴表示不同的用户类型，如官方账户（Official）、团队账户（Team）、新闻媒体账户（News）、总部账户（Hq）、管理员账户（Admin）以及其他（Other）。
- Y轴表示每种用户类型的推文数量。
- 图表使用`coolwarm`调色板，颜色代表不同的用户类型。
- 图表标题为“基于用户名关键词的用户类型分布”。
- 横轴标签为“用户类型”，纵轴标签为“推文数量”。
- 图表设计紧凑，确保所有元素在画布上清晰可见。


```python
# Checking if there are any columns that might contain user description or similar information
data.columns
```

```python
# Example keywords that might indicate a certain type of account
keywords = ['official', 'team', 'news', 'hq', 'admin']

# Function to categorize usernames based on keywords
def categorize_username(username):
    for keyword in keywords:
        if keyword in username.lower():
            return keyword.capitalize()
    return 'Other'

# Categorizing usernames
data['UserType'] = data['Username'].apply(categorize_username)

# Counting the occurrences of each user type
user_type_counts = data['UserType'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=user_type_counts.index, y=user_type_counts.values, palette="coolwarm")
plt.title('Distribution of User Types Based on Username Keywords')
plt.xlabel('User Type')
plt.ylabel('Number of Tweets')
plt.tight_layout()

# Show the plot
plt.show()
```

# 被提及用户的分析

这个水平条形图展示了数据集中最常被提及的前十个用户。

- Y轴显示了用户名。
- X轴表示每个用户被提及的次数。
- 图表使用`viridis`调色板，颜色从顶部的深色渐变到底部的浅色。
- 图表标题为“最常被提及的前十个用户名”。

```python
# Checking the detailed structure of a few entries in the 'MentionedUsers' column
data['MentionedUsers'].dropna().head()
```

```python
# Extracting mentioned usernames using a regular expression
mentioned_usernames = []

for item in data['MentionedUsers'].dropna():
    # Finding all occurrences of usernames in the string
    found_usernames = re.findall(r"username='(.*?)'", item)
    mentioned_usernames.extend(found_usernames)

# Counting the occurrences of each mentioned username
mentioned_username_counts = Counter(mentioned_usernames)

# Selecting the top 10 mentioned usernames for visualization
top_mentioned_usernames = dict(mentioned_username_counts.most_common(10))

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=list(top_mentioned_usernames.values()), y=list(top_mentioned_usernames.keys()), palette="viridis")
plt.title('Top 10 Most Mentioned Usernames')
plt.xlabel('Number of Mentions')
plt.ylabel('Username')
plt.tight_layout()

# Show the plot
plt.show()
```

# 时间段活跃度分析

此条形图展示了不同时间段内的推文活动数量，以了解用户在一天中的哪些时段最为活跃。

- X轴展示了一天中的不同时间段，例如“早晨”、“下午”、“晚上”和“夜晚”。
- Y轴显示了相应时间段内的推文数量。
- 条形图的颜色随着时间段的不同而变化，使用了`Set2`调色板。
- 根据图表结果，可以看出“早晨”时段的推文数量最多，随后是“下午”、“晚上”和“夜晚”。
- 图表标题为“一天中的推文活动情况”。
- 横轴标签为“时间段”，纵轴标签为“推文数量”。
- 通过`plt.tight_layout()`优化了图表布局，使所有内容在画布上均匀分布。


```python
# Function to categorize time of day
def categorize_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'

# Extracting the hour from the datetime
data['HourOfDay'] = data['Datetime'].dt.hour

# Categorizing tweets by time of day
data['TimeOfDay'] = data['HourOfDay'].apply(categorize_time_of_day)

# Counting the number of tweets in each time category
tweet_count_by_time_of_day = data['TimeOfDay'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=tweet_count_by_time_of_day.index, y=tweet_count_by_time_of_day.values, palette="Set2")
plt.title('Tweet Activity by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Tweets')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Re-importing the dataset and necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import ast

# Loading the dataset
file_path = './data/chatgpt1.csv'  # Ensure the correct file path is used
data = pd.read_csv(file_path)

# Extracting hashtags from each tweet
all_hashtags = []
for hashtags_list in data['hashtag'].dropna().apply(ast.literal_eval):
    all_hashtags.extend(hashtags_list)

# Creating a dataframe for tweets with hashtags
tweets_with_hashtags = data[data['hashtag'].notna()]

# Function to check if a tweet contains a specific hashtag
def contains_hashtag(hashtag_list, hashtag):
    return hashtag in ast.literal_eval(hashtag_list) if hashtag_list else False

# Calculating average interactions for the top hashtags
top_hashtags = [tag for tag, count in Counter(all_hashtags).most_common(5)]

# Initialize an empty list to store average interaction data
average_interaction_data = []

for hashtag in top_hashtags:
    subset = tweets_with_hashtags[tweets_with_hashtags['hashtag'].apply(contains_hashtag, hashtag=hashtag)]
    mean_values = subset[['LikeCount', 'RetweetCount', 'ReplyCount']].mean()
    average_interaction_data.append(mean_values.to_list() + [hashtag])

# Creating a DataFrame from the list
average_interactions = pd.DataFrame(average_interaction_data, columns=['LikeCount', 'RetweetCount', 'ReplyCount', 'Hashtag'])

# Melting the dataframe for better plotting
average_interactions_melted = average_interactions.melt(id_vars='Hashtag', var_name='InteractionType', value_name='AverageCount')

# Plotting
plt.figure(figsize=(15, 8))
sns.barplot(x='Hashtag', y='AverageCount', hue='InteractionType', data=average_interactions_melted, palette='viridis')
plt.title('Average User Interactions by Hashtag')
plt.xlabel('Hashtag')
plt.ylabel('Average Number of Interactions')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

# 社交网络分析

本次分析构建了一个基于用户互动的社交网络图，节点代表用户，边代表互动（回复、转发、点赞）。

- 网络中共有38,840个节点和71,218条边。
- 通过筛选至少有一个非零权重边的节点，创建了一个子图，包含22,473个节点和34,952条边。
- 网络的密度非常低，为9.442185065183267e-05，说明用户间的互动非常稀疏。
- 利用随机游走的方法估计网络的平均最短路径长度，结果为2.206，这意味着在用户间转发一条信息平均需要通过2.206个步骤。

这些统计数据提供了对社交网络结构和信息传播效率的洞察。


```python
import networkx as nx

G = nx.Graph()

df = data

# Defining the nodes
G.add_nodes_from(df['Username'])

# Add edges to the graph based on the UserFollowers and UserFriends columns
for index, row in df.iterrows():
    user = row['Username']
    replys = row['ReplyCount']
    retweets = row['RetweetCount']
    likes = row['LikeCount']
    quote = row['QuoteCount']
    G.add_edge(user, replys, weight=replys)
    G.add_edge(user, retweets, weight=retweets)
    G.add_edge(user, likes, weight=likes)
    #G.add_edge(user, quote, weight=quote)

print(G)
```

```python
# Create a subgraph of nodes that have at least one non-zero edge weight
subgraph_nodes = set()
for u, v, data in G.edges(data=True):
    if data['weight'] > 0:
        subgraph_nodes.add(u)
        subgraph_nodes.add(v)
subgraph = G.subgraph(subgraph_nodes)

print(subgraph)
print("Number of nodes in subgraph:", subgraph.number_of_nodes())
print("Number of edges in subgraph:", subgraph.number_of_edges())
```

```python
def print_network_stats(G):
    # Number of Nodes
    print ("Number of Nodes: ",G.number_of_nodes())
    # Number of Edges
    print ("Number of Edges: ", G.number_of_edges())
    # Network density
    print("Network Density: ", nx.density(G))

print_network_stats(G)
```

```python
import numpy as np
# Set the number of random walks and the length of each walk
num_walks = 1000
walk_length = 100

# Initialize the estimate of the average shortest path length
avg_shortest_path = 0

# Perform random walks with restarts and accumulate the distances
for i in range(num_walks):
    # Start a random walk from a randomly chosen node
    start_node = np.random.choice(list(G.nodes()))
    curr_node = start_node
    
    # Perform the random walk
    for j in range(walk_length):
        if curr_node not in G:
            # If the current node is not in the graph, restart the walk from a random node
            start_node = np.random.choice(list(G.nodes()))
            curr_node = start_node
        else:
            # Choose a random neighbor of the current node
            neighbors = list(G.neighbors(curr_node))
            if len(neighbors) == 0:
                # If the current node has no neighbors, restart the walk from a random node
                start_node = np.random.choice(list(G.nodes()))
                curr_node = start_node
            else:
                curr_node = np.random.choice(neighbors)
    
    # Compute the distance between the start node and the current node
    avg_shortest_path += nx.shortest_path_length(G, start_node, curr_node)

# Normalize the accumulated distances by the number of random walks
avg_shortest_path /= num_walks

print("The Network Average Shortest Path (estimated): ", avg_shortest_path)
```

```python
# Transforming the degrees of centrality to pandas dataframe
dc = nx.degree_centrality(G)

dc_df = pd.DataFrame.from_dict({
    'node': list(dc.keys()),
    'centrality': list(dc.values())
})

dc_df.sort_values('centrality', ascending=False)

# Summary statistics centrality degree
dc_df['centrality'].describe()
```

## 数据预处理
首先，我对数据进行了预处理。我从一个 CSV 文件中读取数据，并将日期时间转换为适当的 datetime 格式。为了更好地分析和建模，我还设置了 datetime 作为数据框架的索引。

然后，我将数据按天重采样，并对每日的“点赞”数进行了累加。这一步是为了简化模型输入，并能更清晰地观察到每日“点赞”数的趋势。

## 模型构建与预测
在数据准备完毕后，我选择了 ARIMA 模型进行拟合，具体参数为 (2, 1, 2)。这组参数意味着模型包含两个自回归项，一个差分项以及两个移动平均项，这通常适用于具有短期波动性的时间序列数据。

在模型拟合完成后，我进行了未来 30 天的“点赞”数预测。这一预测帮助我们理解未来的趋势，并对策略做出相应的调整。

## 结果解释
在结果图中，可以看到实际的每日“点赞”数和预测的“点赞”数。实际数据显示了之前的趋势，而预测数据则给出了未来的可能走势。通过比较这两条曲线，我能够评估模型的预测性能，并对模型进行进一步的调整以提高准确度。


```python
import pandas as pd
from pmdarima import auto_arima
import matplotlib.pyplot as plt

# Reload the dataset
file_path = './data/chatgpt1.csv'
new_data = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the new dataset
new_data.head()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Reload the dataset
new_data = pd.read_csv(file_path)

# Convert 'Datetime' to datetime and set it as index
new_data['Datetime'] = pd.to_datetime(new_data['Datetime'])
new_data.set_index('Datetime', inplace=True)

# Resample data to daily frequency, summing up the LikeCounts
daily_likes = new_data['LikeCount'].resample('D').sum()

# Fit ARIMA model (we'll use p=2, d=1, q=2 as common starting parameters)
model = ARIMA(daily_likes, order=(2, 1, 2))
fit_model = model.fit()

# Forecast the next 30 days
forecast = fit_model.get_forecast(steps=30)
forecast_index = pd.date_range(start=daily_likes.index[-1], periods=31, freq='D')[1:]
forecast_values = forecast.predicted_mean

# Plot the original data and the forecast
plt.figure(figsize=(14, 7))
plt.plot(daily_likes, label='Actual Likes')
plt.plot(forecast_index, forecast_values, label='Forecast Likes', color='red')
plt.title('Likes Time Series Forecast')
plt.xlabel('Date')
plt.ylabel('Total Daily Likes')
plt.legend()
plt.show()
```


# 深度学习在时间序列预测中的应用

## 项目介绍

在这个项目中，我利用深度学习中的长短期记忆网络（LSTM）来预测社交媒体的“点赞”数量。LSTM 是处理序列数据的强大工具，尤其适合时间序列分析。

## 数据处理

我首先读取了 CSV 数据文件，并将时间戳转换为 datetime 格式，设置为数据框架的索引。之后，我对“点赞”数量进行了按日重采样，并填补了任何缺失的数据。

## 数据标准化

考虑到 LSTM 模型的输入需要标准化的数值，我使用了最小最大缩放器（MinMaxScaler）对数据进行了标准化处理，将数据缩放到了 0 到 1 的范围内。

## 模型构建

由于我们的数据集较小，我直接准备了训练数据而没有使用 `TimeseriesGenerator`。我构建了一个包含 50 个 LSTM 单元和一个全连接层的序贯模型，并选择了 `adam` 优化器和均方误差损失函数。

## 训练与预测

模型训练了 200 个周期，我监控了损失函数的下降，以确保模型在学习过程中的表现。最后，我进行了对最后一个数据点的单步预测，并将标准化的预测结果转换回实际的“点赞”数量。

## 结果

模型预测的“点赞”数量为 101788.59，表明了我的模型对未来趋势有着相当的预测能力。


```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
import matplotlib.pyplot as plt

# 加载数据
file_path = './data/chatgpt1.csv'
data = pd.read_csv(file_path)
data['Datetime'] = pd.to_datetime(data['Datetime'])
data.set_index('Datetime', inplace=True)
daily_likes = data['LikeCount'].resample('D').sum().fillna(0)

# 数据标准化
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(daily_likes.values.reshape(-1, 1))

# 创建时间序列数据集
n_input = 1  # 只能使用1天的数据来预测下一天
n_features = 1  # 特征数量为 1

# 由于数据点太少，我们不使用 TimeseriesGenerator
# 直接准备数据
train_X, train_y = scaled_data[:-1], scaled_data[1:]

# 构建 LSTM 模型
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# 由于数据点太少，我们不使用 fit_generator 方法
model.fit(train_X, train_y, epochs=200, verbose=1)

# 进行单步预测
x_input = scaled_data[-1].reshape((1, n_input, n_features))
yhat = model.predict(x_input, verbose=0)

# 反转标准化以得到实际预测值
predicted_like = scaler.inverse_transform(yhat)

print('Predicted Like:', predicted_like)
```

# 决策树回归模型分析

## 项目简介

本项目使用决策树回归模型对社交媒体上的“点赞”数量进行了预测。决策树是一种常用的机器学习算法，适用于回归和分类问题。在这个案例中，我们将其应用于回归问题。

## 数据预处理

首先，我从 CSV 文件中加载数据，然后将日期时间转换为 datetime 格式，并提取出年中的天数作为模型的特征。

## 模型构建与训练

我使用了决策树回归器来拟合数据。决策树通过学习数据中的规律，来预测点赞数与年中天数之间的关系。

## 预测与可视化

在模型训练完成后，我为接下来的 30 天创建了新的数据点，并用训练好的模型进行了预测。我使用散点图展示了实际数据点，并用蓝色线条展示了模型的预测结果。

## 结果分析

通过可视化的结果，我们可以看到模型对实际数据点的拟合程度以及对未来点赞数的预测。


```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 加载数据
file_path = './data/chatgpt1.csv'
data = pd.read_csv(file_path)
data['Datetime'] = pd.to_datetime(data['Datetime'])
data['Day'] = data['Datetime'].dt.dayofyear  # 将日期转换为年中的天数

# 准备数据
X = data['Day'].values.reshape(-1, 1)  # 特征：年中的天数
y = data['LikeCount'].values  # 目标：点赞数

# 创建决策树回归模型并拟合
model = DecisionTreeRegressor()
model.fit(X, y)

# 创建用于预测的新数据点
X_future = np.array(range(min(data['Day']), max(data['Day']) + 30)).reshape(-1, 1)

# 预测
y_future = model.predict(X_future)

# 可视化
plt.scatter(X, y, color='black')
plt.plot(X_future, y_future, color='blue', linewidth=2)
plt.xlabel('Day of Year')
plt.ylabel('Like Count')
plt.title('Decision Tree Regression on Likes')
plt.show()
```
